```python
Cpe=1.2
qz=[0.69,0.96,1.5]
s=3  
L=[]
pn=[]
w=[]
M=[]
for i in range(len(qz)):
    pn.append(Cpe*qz[i])
    w.append(pn[i]*s)
    M.append([])    
    for j in range(12):
        if i==0: L.append(j)
        M[i].append(w[i]*L[j]**2/8)
```

```python
for i in range(3):
    print(["{:5.2f}".format(item) for item in M[i]])
```
